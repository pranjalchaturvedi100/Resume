```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('data_set.csv')
raw_data
```

## Dataset without feature engineering (Orignal_Dataset) for PM 1

```python
# Dropping Id Column
dataset1 = raw_data.drop('id', axis=1)

# Filling Na values with the mean of that column
for column in dataset1.columns[1:]:
    dataset1[column] = dataset1[column].fillna(dataset1[column].mean())

# Replacing target/diagnosis column with 1 and -1
dataset1.diagnosis.replace(['M', 'B'], [1, -1], inplace=True)
dataset1
```

## Dataset with feature engineering (Featured_Dataset) for PM 3

```python
# Dropping Id Column
dataset3 = raw_data.drop('id', axis='columns')

# Normalising Data
for column in dataset3.columns[1:]:
    dataset3[column] = (dataset3[column] - dataset3[column].mean())/dataset3[column].std()

# Replacing target/diagnosis column with 1 and -1
dataset3.diagnosis.replace(['M', 'B'], [1, -1], inplace=True)

# Filling Na values with the mean of that column i.e 0;
dataset3.fillna(0, inplace=True)
dataset3
```

## Dataset formed by rearranging the features of Orignal_Dataset for PM 4

```python
# Rearranging Features
diagnosis = dataset1.diagnosis
dataset4 = dataset1.sample(frac=1, axis=1)
dataset4.drop('diagnosis', axis=1, inplace=True)
dataset4.insert(0, 'diagnosis', diagnosis)
dataset4
```

## Activating Function for Perceptron

```python
def func(x):
    if x>=0:
        return 1
    else:
        return -1
```

## Perceptron Algorithm

```python
def train_perceptron(training_set):
    w = np.zeros(31)
    for i in range(3000):
        for training_example in training_set:
            x = np.append(training_example[1:], 1)
            y = training_example[0]
            if y*func(w.dot(x))<0:
                w = w + y*x  
    return w
```

## Evaluation

```python
def evaluate_perceptron(testing_set, w):
    c_mat = np.zeros((2, 2))
    for testing_example in testing_set:
        x = np.append(testing_example[1:], 1)
        y = testing_example[0]
        if y==1 and func(w.dot(x))==1:
            c_mat[1][1]+=1
            
        elif y==1 and func(w.dot(x))==-1:
            c_mat[1][0]+=1
            
        elif y==-1 and func(w.dot(x))==-1: 
            c_mat[0][0]+=1
            
        else:
            c_mat[0][1]+=1
            
    return c_mat
```

## Evaluation Metrics

```python
def accuracy(c_mat):
    return (c_mat[0][0]+c_mat[1][1])/(c_mat[0][0] + c_mat[0][1] + c_mat[1][0] + c_mat[1][1])*100

def precision(c_mat):
    return (c_mat[0][0])/(c_mat[1][0] + c_mat[0][0])*100

def recall(c_mat):
    return (c_mat[1][1])/(c_mat[1][0] + c_mat[1][1])*100
```

## Spilting the Dataset into Training Data and Testing Data

```python
training_data = dataset1.sample(frac=2/3)
training_data
```

```python
testing_data = dataset1.drop(training_data.index)
testing_data
```

## Training Perceptron using the above mentioned algorithm

```python
training_set = training_data.to_numpy()
w1 = train_perceptron(training_set)
w1
```

## Evaluating the Model on Testing Data

```python
testing_set = testing_data.to_numpy()
n = evaluate_perceptron(testing_set, w1)
print(n)
print(f"Accuracy: {accuracy(n)}")
print(f"Precision: {precision(n)}")
print(f"Recall: {recall(n)}")
```

## Perceptron Model 1

```python
accuracy_list = np.zeros(10)
precision_list = np.zeros(10)
recall_list = np.zeros(10)
training_sets = []
w = np.zeros(31)
for i in range(10):
    
    training_data = dataset1.sample(frac=2/3)
    testing_data = dataset1.drop(training_data.index)
    training_sets.append(training_data)
    training_set = training_data.to_numpy()
    testing_set = testing_data.to_numpy()
    
    w = train_perceptron(training_set)
    c_mat = evaluate_perceptron(testing_set, w)
    
    print(f"Model No.{i+1}")
    print(f"Confusion Matrix:\n{c_mat}")
    print(f"Accuracy = {accuracy(c_mat)}")
    print(f"Precision = {precision(c_mat)}")
    print(f"Recall = {recall(c_mat)}")
    
    accuracy_list[i] =  accuracy(c_mat)
    precision_list[i] = precision(c_mat)
    recall_list[i] =  recall(c_mat)
    print("\n")
    
print(f"Accuracy:\n Mean = {accuracy_list.mean()}  Standard Deviation = {accuracy_list.std()}")
print(f"Precision:\n Mean = {precision_list.mean()}  Standard Deviation = {precision_list.std()}")
print(f"Recall:\n Mean = {recall_list.mean()}  Standard Deviation = {recall_list.std()}")
```
